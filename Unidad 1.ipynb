```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import skewnorm, gamma

# Function to estimate mode from histogram
def estimate_mode(data):
    counts, bins = np.histogram(data, bins=100, density=True)
    bin_centers = (bins[:-1] + bins[1:]) / 2
    mode_index = np.argmax(counts)
    return bin_centers[mode_index]

# -------------------------------
# Right-Skewed (Positive Skew) Distribution - Gamma
# -------------------------------
shape, scale = 2, 2
right_skewed_data = np.random.gamma(shape, scale, size=10000)

mean_right = np.mean(right_skewed_data)
median_right = np.median(right_skewed_data)
mode_right = estimate_mode(right_skewed_data)

# Quartiles
q1_right, q2_right, q3_right = np.percentile(right_skewed_data, [25, 50, 75])

# Plot right-skewed
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
counts, bins, _ = plt.hist(right_skewed_data, bins=100, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Right-Skewed (Positive Skew)\n(Gamma Distribution)')
plt.xlabel('Value')
plt.ylabel('Density')

# Plot mean, median, mode
plt.axvline(mean_right, color='red', linestyle='-', linewidth=2, label=f'Mean = {mean_right:.2f}')
plt.axvline(median_right, color='green', linestyle='-', linewidth=2, label=f'Median = {median_right:.2f}')
plt.axvline(mode_right, color='blue', linestyle='-', linewidth=2, label=f'Mode ≈ {mode_right:.2f}')

# Plot quartiles
plt.axvline(q1_right, color='orange', linestyle='--', linewidth=2, label=f'Q1 (25th %ile) = {q1_right:.2f}')
plt.axvline(q2_right, color='green', linestyle='--', linewidth=2, label=f'Q2 (50th %ile) = {q2_right:.2f}')  # Same as median
plt.axvline(q3_right, color='purple', linestyle='--', linewidth=2, label=f'Q3 (75th %ile) = {q3_right:.2f}')

plt.legend(fontsize=9)
plt.grid(True, alpha=0.3)

# -------------------------------
# Left-Skewed (Negative Skew) Distribution - Skew-Normal
# -------------------------------
a = -5
left_skewed_data = skewnorm.rvs(a, size=10000)

mean_left = np.mean(left_skewed_data)
median_left = np.median(left_skewed_data)
mode_left = estimate_mode(left_skewed_data)

# Quartiles
q1_left, q2_left, q3_left = np.percentile(left_skewed_data, [25, 50, 75])

# Plot left-skewed
plt.subplot(1, 2, 2)
counts, bins, _ = plt.hist(left_skewed_data, bins=100, density=True, alpha=0.7, color='salmon', edgecolor='black')
plt.title('Left-Skewed (Negative Skew)\n(Skew-Normal Distribution)')
plt.xlabel('Value')
plt.ylabel('Density')

# Plot mean, median, mode
plt.axvline(mean_left, color='red', linestyle='-', linewidth=2, label=f'Mean = {mean_left:.2f}')
plt.axvline(median_left, color='green', linestyle='-', linewidth=2, label=f'Median = {median_left:.2f}')
plt.axvline(mode_left, color='blue', linestyle='-', linewidth=2, label=f'Mode ≈ {mode_left:.2f}')

# Plot quartiles
plt.axvline(q1_left, color='orange', linestyle='--', linewidth=2, label=f'Q1 (25th %ile) = {q1_left:.2f}')
plt.axvline(q2_left, color='green', linestyle='--', linewidth=2, label=f'Q2 (50th %ile) = {q2_left:.2f}')
plt.axvline(q3_left, color='purple', linestyle='--', linewidth=2, label=f'Q3 (75th %ile) = {q3_left:.2f}')

plt.legend(fontsize=9)
plt.grid(True, alpha=0.3)

# Final layout
plt.tight_layout()
plt.show()

# Print skewness and quartile ranges
print(f"Right-skewed data - Skewness: {stats.skew(right_skewed_data):.3f}")
print(f"Right-skewed data - IQR (Q3 - Q1): {q3_right - q1_right:.3f}")
print(f"Left-skewed data - Skewness: {stats.skew(left_skewed_data):.3f}")
print(f"Left-skewed data - IQR (Q3 - Q1): {q3_left - q1_left:.3f}")
```
